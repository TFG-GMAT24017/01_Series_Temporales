```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('once')
```

```python
datos_oferta = pd.read_csv('data/datos_Oferta_Renovables_EU27_GWh.csv', index_col=0, parse_dates=True)['EU27']
# poner el índice con frecuencia diaria
serie_oferta = datos_oferta.asfreq('D')
```

```python
datos_demanda = pd.read_csv('data/datos_Demanda_EU27_GWh.csv', index_col=0, parse_dates=True)['EU27']
# poner el índice con frecuencia diaria
serie_demanda = datos_demanda.asfreq('D')
```

```python
# Alinear temporalmente las dos series
serie_oferta, serie_demanda = serie_oferta.align(serie_demanda, join='inner')

# Graficar ambas series
plt.figure(figsize=(30, 8))
plt.plot(serie_demanda.index, serie_demanda.values, label="Demanda Total UE", color="blue")
plt.plot(serie_oferta.index, serie_oferta.values, label="Oferta Renovables Total UE", color="red")

# Etiquetas y título
plt.xlabel("Fecha")
plt.ylabel("GWh")
plt.legend()
plt.grid()

plt.show()
```
